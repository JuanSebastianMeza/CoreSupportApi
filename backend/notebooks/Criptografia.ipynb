```python
input_key = open('../api/api/private.key').read()
print(input_key)
input_key2 = open('../api/api/public.key').read()
print(input_key2)
```

```python
import jwt
encoded = jwt.encode({'some': 'payload'}, 'secret', algorithm='HS256')
encoded
```

```python
jwt.decode(encoded, 'secret', algorithms=['HS256'])
```

```python
key_private = "\n".join([l.lstrip() for l in input_key.split("\n")])
print(key_private)
key_public = "\n".join([l.lstrip() for l in input_key2.split("\n")])
print(key_public)
key_public
```

```python
encoded_rsa = jwt.encode(
    {'some': 'payload'},
    key_private,
    algorithm='RS256'
)
encoded_rsa
```

```python
jwt.decode(
    encoded_rsa,
    key_public,
    algorithm='RS256'
)
```
