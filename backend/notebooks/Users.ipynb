# Create new user

```python
from django.contrib.auth.models import User
from django.contrib.auth.hashers import make_password
from authentication.models import Profile, Position, Department, PasswordHistory

# New user data
password='abc123++'
is_superuser=False
username='E01098'
first_name='Luis'
last_name='Rinaldi'
email='luis.rinaldi@telefonica.com'
is_staff=False
position=2
department=5

# Create user
new_user = User(username=username, first_name=first_name, last_name=last_name, email=email)
new_user.is_superuser = is_superuser
new_user.is_staff = is_staff
new_user.set_password(password)
new_user.save()
print(new_user)

# Create Profile
new_profile = Profile(user=new_user, position=Position.objects.get(id=position), department=Department.objects.get(id=department))
new_profile.save()
print(new_profile)

# Add new password to history
user_history = PasswordHistory(user=new_user, password=make_password(password))
user_history.save()
print(user_history)
```
